### __Research Question 3: 这种 disparity 是否和当地的社会经济属性存在关联__

- We need to calculate the Variance Inflation Factor (VlF) before choosing the model. 

- In PLS, the coefficients for each predictor are derived based on the components, but the p-values for each original variable are not directly computed because PLS does not operate like traditional OLS regression on the raw predictors. If we still want to use PLS and get p-values for each original variable, we can run permutation tests to generate p-values by repeatedly permuting the target variable and recalculating the PLS coefficients.

```python
# Extract independent variables and perform standardization
X = df_merged[['Democrat_R', 'Republican_R', 'Total_Population', 'Median_income', 'GINI', 'No_Insurance_R', 
               'Household_Below_Poverty_R', 'HISPANIC_LATINO_R', 'White_R', 'Black_R', 'Indian_R', 'Asian_R', 
               'Under_18_R', 'Bt_18_44_R', 'Bt_45_64_R', 'Over_65_R', 'Male_R', 'Bachelor_R',
               'Education_Degree_R', 'Population_Density', 'Unemployed_R']].fillna(0)

scaler_X = StandardScaler()
X_scaled = scaler_X.fit_transform(X)

# Create DataFrame after standardization
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

# Function to calculate VIF
def calculate_vif(X):
    vif_data = pd.DataFrame()
    vif_data["Feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data

# Calculate and display VIF
vif_data = calculate_vif(X_scaled_df)

vif_data
```

```python
# significant multicollinearity (vif > 5) so we will use pls

# Define independent (X) and dependent variables (y) for each period
X = df_merged[['Democrat_R', 'Republican_R', 'Total_Population', 'Median_income', 'GINI', 'No_Insurance_R', 
               'Household_Below_Poverty_R', 'HISPANIC_LATINO_R', 'White_R', 'Black_R', 'Indian_R', 'Asian_R', 
               'Under_18_R', 'Bt_18_44_R', 'Bt_45_64_R', 'Over_65_R', 'Male_R', 'Bachelor_R',
               'Education_Degree_R', 'Population_Density', 'Unemployed_R']].fillna(0)


y_pre = df_merged['Pre-COVID_average_sentiment'].fillna(0)
y_during = df_merged['During-COVID_average_sentiment'].fillna(0)
y_post = df_merged['Post-COVID_average_sentiment'].fillna(0)

# Standardize independent variables
X_scaled = StandardScaler().fit_transform(X)

# root_mean_squared_error instead od mean_squared_error
def pls_regression_with_cv(X, y, n_permutations=1000, n_splits=5):
    pls = PLSRegression(n_components=2)
    pls.fit(X, y)
    coefficients = pls.coef_.ravel()
    
    # R² and RMSE calculation
    y_pred = pls.predict(X)
    r2 = r2_score(y, y_pred)
    rmse = root_mean_squared_error(y, y_pred)
    
    # Cross-validation setup
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
    r2_cv_scores = []
    rmse_cv_scores = []
    
    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        
        pls.fit(X_train, y_train)
        y_pred_cv = pls.predict(X_test)
        
        r2_cv_scores.append(r2_score(y_test, y_pred_cv))
        rmse_cv_scores.append(root_mean_squared_error(y_test, y_pred_cv))
    
    r2_cv_mean = np.mean(r2_cv_scores)
    rmse_cv_mean = np.mean(rmse_cv_scores)
    
    # Permutation testing
    p_values, std_errors = [], []
    for i in range(X.shape[1]):
        original_score = np.corrcoef(X[:, i], y.ravel())[0, 1]
        permuted_scores = [np.corrcoef(X[:, i], resample(y).ravel())[0, 1] for _ in range(n_permutations)]
        p_values.append(np.sum(np.abs(permuted_scores) >= np.abs(original_score)) / n_permutations)
        std_errors.append(np.std(permuted_scores))
    
    return coefficients, p_values, std_errors, r2, rmse, r2_cv_mean, rmse_cv_mean


# Run PLS regression for each period
results = {}
for period, y in zip(['Pre', 'During', 'Post'], [y_pre, y_during, y_post]):
    coeffs, pvals, stderrs, r2, rmse, r2_cv, rmse_cv = pls_regression_with_cv(X_scaled, y.values.reshape(-1, 1))
    results[period] = (coeffs, pvals, stderrs)
    
    # Print R² and RMSE for each period
    print(f'{period}-Pandemic R²: {r2}, CV R²: {r2_cv}')
    print(f'{period}-Pandemic RMSE: {rmse}, CV RMSE: {rmse_cv}')

# Create DataFrames for each period
columns = X.columns
df_pre = pd.DataFrame({'Coefficient': results['Pre'][0], 'P-Value': results['Pre'][1], 
                       'Standard Error': results['Pre'][2]}, index=columns)

df_during = pd.DataFrame({'Coefficient': results['During'][0], 'P-Value': results['During'][1], 
                          'Standard Error': results['During'][2]}, index=columns)

df_post = pd.DataFrame({'Coefficient': results['Post'][0], 'P-Value': results['Post'][1], 
                        'Standard Error': results['Post'][2]}, index=columns)

# Concatenate results
df_final = pd.concat([df_pre.add_prefix('Pre-Pandemic '), 
                      df_during.add_prefix('During-Pandemic '), 
                      df_post.add_prefix('Post-Pandemic ')], axis=1)

# Display the final results
df_final
```

```python
# During - Before
y_diff = df_merged['During-COVID_average_sentiment'].fillna(0) - df_merged['Pre-COVID_average_sentiment'].fillna(0)

# Function to perform PLS regression with cross-validation and permutation testing
def pls_regression_with_cv_and_permutation(X, y, n_permutations=1000, n_splits=5):
    pls = PLSRegression(n_components=2)
    pls.fit(X, y)
    coefficients = pls.coef_.ravel()

    # R² and RMSE calculation
    y_pred = pls.predict(X)
    r2 = r2_score(y, y_pred)
    rmse = root_mean_squared_error(y, y_pred)
    
    # Cross-validation setup
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
    r2_cv_scores = []
    rmse_cv_scores = []
    
    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        
        pls.fit(X_train, y_train)
        y_pred_cv = pls.predict(X_test)
        
        r2_cv_scores.append(r2_score(y_test, y_pred_cv))
        rmse_cv_scores.append(root_mean_squared_error(y_test, y_pred_cv))
    
    r2_cv_mean = np.mean(r2_cv_scores)
    rmse_cv_mean = np.mean(rmse_cv_scores)
    
    # Permutation testing
    p_values = []
    std_errors = []
    
    for i in range(X.shape[1]):
        original_score = np.corrcoef(X[:, i], y.ravel())[0, 1]
        permuted_scores = [np.corrcoef(X[:, i], resample(y).ravel())[0, 1] for _ in range(n_permutations)]
        p_value = (np.sum(np.abs(permuted_scores) >= np.abs(original_score))) / n_permutations
        std_error = np.std(permuted_scores)
        p_values.append(p_value)
        std_errors.append(std_error)
    
    return coefficients, p_values, std_errors, r2, rmse, r2_cv_mean, rmse_cv_mean

# Run PLS regression using the difference between during-pandemic and pre-pandemic sentiment as y
coeff_diff, pval_diff, stderr_diff, r2_diff, rmse_diff, r2_cv_diff, rmse_cv_diff = pls_regression_with_cv_and_permutation(X_scaled, y_diff.values.reshape(-1, 1))

# Print R² and RMSE results
print(f'During-Before R²: {r2_diff}, CV R²: {r2_cv_diff}')
print(f'During-Before RMSE: {rmse_diff}, CV RMSE: {rmse_cv_diff}')

# Create a DataFrame to hold the results
columns = X.columns
df_diff = pd.DataFrame({'Coefficient': coeff_diff, 'P-Value': pval_diff, 'Standard Error': stderr_diff}, index=columns)

# Display the final results for the difference
df_diff
```

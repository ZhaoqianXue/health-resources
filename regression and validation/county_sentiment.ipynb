- 疫情时: 2020.2.1-2020.5.31 (大部分州恢复秩序)
- 疫情后: 2020.6.1-2021.5.31 (新变种德尔塔爆发以前)

```python
import pandas as pd
import os
import geopandas as gpd
from shapely.geometry import Point
```

```python
# Data of all states
folder_path = "./store-meta"

# List to hold all the state dfs
state_dataframes = []

# Loop through each file in the folder and append the df to the list
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        file_path = os.path.join(folder_path, file_name)
        df_state = pd.read_csv(file_path)
        state_dataframes.append(df_state)

# Combine all state dfs into one df
df_states = pd.concat(state_dataframes, ignore_index=True)
df_states
```

```python
# Dropping duplicate rows
df_states = df_states.drop_duplicates()
df_states
```

```python
# Keep only useful columns for df_states
columns_to_keep_states = ['gmap_id', 'latitude', 'longitude']
df_states = df_states[columns_to_keep_states]
df_states
```

```python
# Load the shapefile for states
states_gdf = gpd.read_file("./state shapefile/cb_2018_us_state_5m.shp")

# Ensure the CRS of states_gdf is EPSG:4326 for consistency
states_gdf = states_gdf.to_crs(epsg=4326)

# Convert df_states to a Geo-df by creating a geometry column from latitude and longitude
df_states = df_states.copy()  # Make a copy to avoid modifying the original df
df_states['geometry'] = df_states.apply(lambda row: Point(row['longitude'], row['latitude']), axis=1)
df_states_gdf = gpd.GeoDataFrame(df_states, geometry='geometry', crs="EPSG:4326")

# Ensure consistent data types for merging
df_states_gdf['gmap_id'] = df_states_gdf['gmap_id'].astype(str)

# Perform spatial intersection using overlay
df_states_with_state = gpd.overlay(df_states_gdf, states_gdf, how='intersection')

# Assign state names and keep only the relevant columns
df_states_with_state['state'] = df_states_with_state['NAME']
df_states = df_states_with_state[['gmap_id', 'latitude', 'longitude', 'state']]

# Final output
df_states
```

```python
# Load the shapefile for counties
county_gdf = gpd.read_file('./county shapefile/tl_2023_us_county.shp')

# Ensure the CRS of county_gdf is EPSG:4326 for consistency
county_gdf = county_gdf.to_crs(epsg=4326)

# Convert df_merged to a Geo-df by creating a geometry column from latitude and longitude
df_states = df_states.copy()  # Ensure we are working on a copy of the df
df_states.loc[:, 'geometry'] = df_states.apply(lambda row: Point(row['longitude'], row['latitude']), axis=1)
df_merged_gdf = gpd.GeoDataFrame(df_states, geometry='geometry', crs="EPSG:4326")

# Perform spatial join between df_merged and the county shapefile using predicate="within"
df_merged_with_county = gpd.sjoin(df_merged_gdf, county_gdf, how="left", predicate="within")

# Add the county column
df_merged_with_county['county'] = df_merged_with_county['NAME']

# Drop the geometry column for final output
df_merged_with_county.drop(columns='geometry', inplace=True)

# Keep only the useful columns
columns_to_keep_county = ['gmap_id', 'county', 'state']
df_county = df_merged_with_county[columns_to_keep_county]

df_county
```

```python
# Load the review (health resource-related) data
df_review = pd.read_csv('./classed_review_2018.csv')
df_review
```

```python
# Keeping only the specified columns
columns_review = ['gmap_id', 'text', 'time', 'class']
df_review = df_review[columns_review]
df_review
```

```python
# Dropping duplicate rows from df_reviews
df_review = df_review.drop_duplicates()
df_review
```

```python
# Merge both dfs based on 'gmap_id' column
df_merged = pd.merge(df_county, df_review, on='gmap_id', how='inner')
df_merged
```

```python
# Convert the time column to datetime format
df_merged['time'] = pd.to_datetime(df_merged['time'])

# Define the time periods based on events
periods = {
    'Pre-COVID': ('2018-01-01', '2020-01-31'),
    'During-COVID': ('2020-02-01', '2020-05-31'),
    'Post-COVID': ('2020-06-01', '2021-05-31')
}

# Function to calculate the review count and average sentiment for each period
def calculate_sentiment(df, start_date, end_date):
    period_df = df[(df['time'] >= start_date) & (df['time'] <= end_date)]
    grouped_df = period_df.groupby(['county', 'state']).agg(
        review_count=('gmap_id', 'size'),  # count the number of reviews
        total_sentiment=('class', lambda x: (x == 1).sum() - (x == -1).sum())  # calculate the total sentiment
    ).reset_index()
    grouped_df['average_sentiment'] = grouped_df['total_sentiment'] / grouped_df['review_count']
    return grouped_df

# Create empty df to store the combined results
result_df = df_merged[['county', 'state']].drop_duplicates().copy()

# Loop through periods and add the review count and average sentiment for each period
for period, (start, end) in periods.items():
    period_data = calculate_sentiment(df_merged, start, end)
    period_data.columns = ['county', 'state', f'{period}_review_count', f'{period}_total_sentiment', f'{period}_average_sentiment']
    result_df = pd.merge(result_df, period_data[['county', 'state', f'{period}_review_count', f'{period}_average_sentiment']], on=['county', 'state'], how='left')

# Display the final df
result_df
```
